```python
def is_valid_password(password):
    if len(password) < 6:
        return False

    has_upper = False
    has_lower = False
    has_digit = False
    has_special = False

    for ch in password:
        if ch.isupper():
            has_upper = True
        elif ch.islower():
            has_lower = True
        elif ch.isdigit():
            has_digit = True
        else:
            has_special = True

    if has_upper and has_lower and has_digit and has_special:
        return True
    else:
        return False


pwd = input("Enter your password: ")

if is_valid_password(pwd):
    print("Correct form")
else:
    print("Wrong format password")
```
