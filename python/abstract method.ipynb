```python
#abstract method
from abc import ABC, abstractmethod
```

```python
class employee(ABC):
    @abstractmethod
    def calculate_salary(self):
        pass
    @abstractmethod
    def details(self):
        pass
```

```python
class employees(employee):
    def __init__(self):
        self.name=input('enter name')
        self.salary=int(input('enter salary'))
        self.pf=self.salary*0.20
        self.calculate_salary()
    def calculate_salary(self):
        self.net_salary=self.salary+self.pf
        print(self.net_salary)
    def details(self):
        print (f"name of employee is {self.name} and its salary is {self.net_salary}")
```

```python
user1=employees()
```

```python
try:
    a = int(input('num1: '))
    b = int(input('num2: '))
    print(a / b)

except ZeroDivisionError:
    print('number cannot divide by zero')

except ValueError:
    print('enter number only')

finally:
    print('program ends successfully')
```

```python
try:
    a=int(input('num1'))
except (ValueError):
    print('taking default number')
try:
    b=int(input('num2'))
except (ValueError):
    print('taking default number')
try:
    print(a/b)
except (ZeroDivisionError):
    print('number cannot divide by zero')
finally:
    print("sum",a+b)
    print("mul",a*b)
```
