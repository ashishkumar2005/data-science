```python
expenses=np.array([
    [200,340,230,150,456],  #food
    [100,345,122,600,233],  #travel
    [350,450,256,800,654]   #shopping
])  
```

```python
print('total expense: ',np.sum(expenses))
print('category wise sum: ',np.sum(expenses,axis=1))
print('day wise total: ',np.sum(expenses,axis=0))
print('highset expense day:',np.max(np.sum(expenses,axis=0)))
```
