```python
#using __init__()
class car:
    model='BMW'
    def __init__(self,year,color):
        self.year=year
        self.color=color

car1=car(2022,'blue')
print(car1.year)
print(car1.color)
```

```python
#use __str__() to make string redable.
class Dog:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def __str__(self):
        return f"{self.name} is {self.age} years old."
dog1 = Dog("Buddy", 3)
dog2 = Dog("Charlie", 5)

print(dog1)  
print(dog2)
```

```python
class student:
    course='B.Tech'
    def __init__(self,name,roll):
        self.name=name
        self.roll=roll

std1=student('alex',223576)
std2=student('joe',223577)

print(std1.roll)

#modifying student1 roll 
std1.roll=223578
print("after modifying: ",std1.roll)

#calling class and instance variable
print(student.course) #class vriable
print(std2.name)      #instance variable
print(std1.name)      #instance variable
```
