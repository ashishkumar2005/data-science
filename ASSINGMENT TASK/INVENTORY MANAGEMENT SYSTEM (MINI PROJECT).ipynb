```python
import os

FILE_NAME = "inventory.txt"

class Product:

    def __init__(self, pid, name, qty, price):
        self.pid = pid
        self.name = name
        self.qty = qty
        self.price = price

    def display(self):
        print(f"{self.pid} | {self.name} | Qty: {self.qty} | Price: ₹{self.price}")

def add_product():
    pid = input("Enter Product ID: ")
    name = input("Enter Product Name: ")
    qty = int(input("Enter Quantity: "))
    price = float(input("Enter Price: "))

    with open(FILE_NAME, "a") as f:
        f.write(f"{pid},{name},{qty},{price}\n")

    print("Product Added Successfully.\n")


def view_products():
    if not os.path.exists(FILE_NAME):
        print("No inventory found.\n")
        return

    print("\n--- Inventory List ---")
    with open(FILE_NAME, "r") as f:
        for line in f:
            pid, name, qty, price = line.strip().split(",")
            product = Product(pid, name, qty, price)
            product.display()
    print()


def search_product():
    search_id = input("Enter Product ID to Search: ")
    found = False

    with open(FILE_NAME, "r") as f:
        for line in f:
            pid, name, qty, price = line.strip().split(",")
            if pid == search_id:
                product = Product(pid, name, qty, price)
                print("\nProduct Found:")
                product.display()
                found = True
                break

    if not found:
        print("Product Not Found.\n")


def update_product():
    pid_update = input("Enter Product ID to Update: ")
    updated = False
    lines = []

    with open(FILE_NAME, "r") as f:
        lines = f.readlines()

    with open(FILE_NAME, "w") as f:
        for line in lines:
            pid, name, qty, price = line.strip().split(",")

            if pid == pid_update:
                print("Enter New Details:")
                name = input("New Name: ")
                qty = input("New Quantity: ")
                price = input("New Price: ")
                updated = True

            f.write(f"{pid},{name},{qty},{price}\n")

    if updated:
        print("Product Updated Successfully.\n")
    else:
        print("Product Not Found.\n")


def delete_product():
    pid_delete = input("Enter Product ID to Delete: ")
    deleted = False
    lines = []

    with open(FILE_NAME, "r") as f:
        lines = f.readlines()

    with open(FILE_NAME, "w") as f:
        for line in lines:
            pid, name, qty, price = line.strip().split(",")

            if pid != pid_delete:
                f.write(line)
            else:
                deleted = True

    if deleted:
        print("Product Deleted Successfully.\n")
    else:
        print("Product Not Found.\n")

def menu():

    while True:
        print("====== INVENTORY MANAGEMENT SYSTEM ======")
        print("1. Add Product")
        print("2. View Products")
        print("3. Search Product")
        print("4. Update Product")
        print("5. Delete Product")
        print("6. Exit")

        choice = input("Enter Choice: ")

        if choice == "1":
            add_product()

        elif choice == "2":
            view_products()

        elif choice == "3":
            search_product()

        elif choice == "4":
            update_product()

        elif choice == "5":
            delete_product()

        elif choice == "6":
            print("Exiting Program...")
            break

        else:
            print("Invalid Choice.\n")

menu()
```
